The overview field contains a description for each movie, so I am gonna work with this description along all movies 

```
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf= TfidfVectorizer(stop_words= "english")
movies_overview=movies["overview"].fillna("")
```

```
tfidf_matrix=tfidf.fit_transform(movies["overview"])
tfidf_matrix
```

### Similarity matrix

```
from sklearn.metrics.pairwise import linear_kernel
similarity_matrix=linear_kernel(tfidf_matrix, tfidf_matrix)
similarity_matrix
```

### Finding the most similar movies to a certain movie

```
movie_title= "John Carter"
```

```
sorted(similarity_matrix[1], reverse= True)
```

```
#we need to make the index match the movie get_feature_names, and not hotcode it
idx=movies.loc[movies["title"]==movie_title].index[0]
idx
#This returns the row with movie title "xxx" and we apply index[0] to give us the index
```

```
#Now we can call idx into our similarity tfidf_matrix
# sorted(similarity_matrix[idx], reverse=True)
```

```
#But nw we dont know which coefficient points to which movie, so we enumerate
scores= list(enumerate(similarity_matrix[idx]))
scores
```

```
# sorted(scores, reverse=True)
# This sorts by the index but we want it sorted by the similarity. findng items with index 1 in each turple
scores=sorted(scores, key=lambda x: x[1], reverse=True)
scores
# movie with index 0 has 3 most similar movies. now lets extract the movie indexes
```

```
movie_indices= [tpl[0] for tpl in scores[1:4]]
movie_indices
# def similar_movies(movie_title, no_movies):
    
```

```
# Now we have the indices we need to extract the title using iloc
list(movies["title"].iloc[movie_indices])
```

```
def similar_movies(movie_title, numb_movies):
    idx=movies.loc[movies["title"]==movie_title].index[0]
    scores= list(enumerate(similarity_matrix[idx]))
    scores=sorted(scores, key=lambda x: x[1], reverse=True)
    movie_indices= [tpl[0] for tpl in scores[1:numb_movies+1]]
    similar_titles=list(movies["title"].iloc[movie_indices])
    return similar_titles
```

```
similar_movies("Furious 7", 2)
```
